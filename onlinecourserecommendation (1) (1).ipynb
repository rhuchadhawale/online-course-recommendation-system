### CONTENT BASED RECOMMENDATION

```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
from scipy.sparse import hstack

import matplotlib.pyplot as plt
```

```python
course_df = df.drop_duplicates(subset='course_name')
```

```python
course_df = df.drop_duplicates(subset='course_name').reset_index(drop=True)
```

```python
print(course_df['course_name'].nunique())
```

```python
course_df['course_name'] = course_df['course_name'].fillna('Unknown')

numeric_cols = [
    'course_price',
    'avg_course_rating',
    'study_material_binary',
    'difficulty_encoded',
    'enrollment_numbers'
]

course_df[numeric_cols] = course_df[numeric_cols].fillna(0)
```

```python
tfidf = TfidfVectorizer(stop_words='english')
course_text = tfidf.fit_transform(course_df['course_name'])
```

```python
numeric_features = course_df[
    [
        'course_price',
        'avg_course_rating',
        'study_material_binary',
        'difficulty_encoded',
        'enrollment_numbers'
    ]
]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numeric_scaled = scaler.fit_transform(numeric_features)
```

```python
from scipy.sparse import hstack

content_matrix = hstack([
    course_text,
    numeric_scaled
]).tocsr()
```

```python
content_matrix = content_matrix.tocsr()
```

```python
from sklearn.neighbors import NearestNeighbors

content_model = NearestNeighbors(
    metric='cosine',
    algorithm='brute'
)

content_model.fit(content_matrix)
```

```python
print(course_df.shape)
print(content_matrix.shape)
```

### recommendation function

```python
def recommend_courses(course_name, top_n=5):

    matches = course_df[
        course_df['course_name'] == course_name
    ].index

    if len(matches) == 0:
        return "Course not found"

    idx = matches[0]

    distances, indices = content_model.kneighbors(
        content_matrix[idx],
        n_neighbors=min(top_n + 1, len(course_df))
    )
    recommendations = course_df.iloc[indices[0][1:]]


    return recommendations[
        [
            'course_name',
            'avg_course_rating',
            'course_price',
            'difficulty_level'
        ]
    ]
```

### Collaborative Filtering

```python
course_user = df.pivot_table(
    index='course_name',
    columns='user_id',
    values='rating',
    fill_value=0
)

print(course_user.shape)
```

```python
course_user.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

course_similarity = cosine_similarity(course_user)
```

 ### course similarity

```python
course_similarity_df = pd.DataFrame(
    course_similarity,
    index=course_user.index,
    columns=course_user.index
)

print(course_similarity_df.shape)
```

```python
course_similarity_df.head()
```

### similarity function

```python
def similar_courses(course_name, top_n=5):

    if course_name not in course_similarity_df.index:
        return "Course not found"

    similar = (
        course_similarity_df.loc[course_name]
        .sort_values(ascending=False)
        .iloc[1:top_n+1]
    )

    result = pd.DataFrame({
        'Course': similar.index,
        'Similarity Score': similar.values
    })

    return result
```

### hybrid function

```python
def hybrid_recommend(course_name, top_n=5):

    # Content-based recommendations
    content_recs = recommend_courses(course_name, top_n)

    content_list = list(content_recs['course_name'])

    # Item-based recommendations
    item_recs = similar_courses(course_name, top_n)

    item_list = list(item_recs['Course'])

    # Merge and remove duplicates
    combined = []

    for course in content_list + item_list:
        if course not in combined:
            combined.append(course)

    return pd.DataFrame({
        'Recommended Course': combined[:top_n]
    })
```

```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

def precision_at_k(recommended, relevant, k=5):
    return len(set(recommended[:k]) & set(relevant)) / k

def recall_at_k(recommended, relevant, k=5):
    if len(relevant) == 0:
        return 0
    return len(set(recommended[:k]) & set(relevant)) / len(relevant)

def f1_score_at_k(p, r):
    if p + r == 0:
        return 0
    return 2 * p * r / (p + r)

results = []

courses_to_test = course_df['course_name'].sample(
    min(50, len(course_df)),
    random_state=42
)

for model_name in ["Content", "Item", "Hybrid"]:

    precisions = []
    recalls = []
    f1s = []

    for course in courses_to_test:

        try:

            # Generate recommendations
            if model_name == "Content":

                recommended = list(
                    recommend_courses(course)['course_name']
                )

            elif model_name == "Item":

                recommended = list(
                    similar_courses(course)['Course']
                )

            else:

                recommended = list(
                    hybrid_recommend(course)['Recommended Course']
                )

            # Get selected course details
            selected_course = course_df[
                course_df['course_name'] == course
            ]

            selected_difficulty = selected_course[
                'difficulty_level'
            ].values[0]

            # Relevant courses = same difficulty level
            relevant = list(
                course_df[
                    course_df['difficulty_level']
                    == selected_difficulty
                ]['course_name']
            )

            # Calculate metrics
            p = precision_at_k(
                recommended,
                relevant,
                k=5
            )

            r = recall_at_k(
                recommended,
                relevant,
                k=5
            )

            f1 = f1_score_at_k(
                p,
                r
            )

            precisions.append(p)
            recalls.append(r)
            f1s.append(f1)

        except Exception:
            pass

    results.append([
        model_name,
        np.mean(precisions),
        np.mean(recalls),
        np.mean(f1s)
    ])

evaluation_df = pd.DataFrame(
    results,
    columns=[
        "Model",
        "Precision@5",
        "Recall@5",
        "F1 Score"
    ]
)

print(evaluation_df)
```

```python
import matplotlib.pyplot as plt

evaluation_df.set_index("Model").plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Recommendation Model Comparison")
plt.ylabel("Score")
plt.xticks(rotation=0)

plt.savefig(
    "model_comparison.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

### CONCLUSION

Final Conclusion

Content-Based Filtering recommends courses based on course features.

Item-Based Collaborative Filtering recommends courses based on user rating patterns.

The Hybrid Recommendation System combines both methods and provides the most comprehensive recommendations for learners. This model was selected as the final recommendation approach because it effectively utilizes both course information and user behavior.
